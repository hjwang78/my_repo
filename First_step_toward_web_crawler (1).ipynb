## 2. Get specific contents using BeatifulSoup

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(data, 'html.parser')
```

```python
# Verifying tables and their classes
print('Classes of each table:')
for table in soup.find_all('table'):
    print(table.get('class'))
```

```python
# Creating list with all tables
tables = soup.find_all('table')

#  Looking for the table with the classes 'wikitable', 'sortable'
table = soup.find('table', class_='wikitable sortable selected_now')
```

```python
table.tbody.find_all('tr')
```

```python
import pandas as pd
# Defining of the dataframe
df = pd.DataFrame(columns=['City','Province','Preference'])

# Collecting Ddata
for row in table.tbody.find_all('tr'):    
    # Find all data for each column
    columns = row.find_all('td')
    
    if (columns != []):
        City = columns[0].text.strip()
        Province = columns[2].text.strip()
        Preference = columns[3].text.strip()
        df = df.append({'City':City, 'Province':Province,'Preference':Preference},ignore_index=True)
```
